```python
import itertools as it
import numpy as np
```

```python
# random

import itertools as it

# all possible Pauliwords on n_qubits!
n_qubits = 3

ham={}
for pauli_comb in it.product(['X', 'Y', 'Z', 'I'],  repeat=n_qubits):
    key = ('').join(pauli_comb)
    ham[key] = np.random.rand(1)

print(len(ham) == 4**n_qubits)
ham
```

```python
test1 = HamiltonianGraph(ham)
cliques_out = test1.clique_cover('C',
                                 'largest_first',
                                 plot_graph=True, 
                                 with_node_label=True, node_sizes=False)

for key in cliques_out:
    print(cliques_out[key])
    print('**')
```

```python
test2 = HamiltonianGraph(ham)
cliques_out2 = test2.clique_cover('AC',
                                 'largest_first',
                                 plot_graph=True, 
                                 with_node_label=True, node_sizes=False)
```
